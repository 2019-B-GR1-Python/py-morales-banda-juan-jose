```python
class Cine:
    ciudad = None
    nombre = None
    sala = 1
    precio = 0
    
    def __init__(self, ciudad = None, nombre = None, sala = 1, precio = 0):
        self.ciudad = ciudad
        self.nombre = nombre
        self.sala = sala
        self.precio = precio
          
    def guardar(self):
        try:
            self.ciudad = (input("Ciudad: "))
            self.nombre = (input("Nombre: "))
            self.sala = (input("Sala: "))
            self.precio = (input("Precio: "))
            path = "./cines.txt"       
            archivo_escritura_abierto = open(path, mode = "a")
            archivo_escritura_abierto.writelines([f"{self.ciudad}\n", 
                                                          f"{self.nombre}\n",
                                                          f"{self.sala}\n",
                                                          f"{self.precio}\n\n"])
            archivo_escritura_abierto.close()   
        except Exception as error:
            print("Error")
        return "Cine Guardado Exitosamente"
        
    def leer(self):
        try:
            path = "./cines.txt"
            archivo_lectura_abierto = open(path)
            contenido = archivo_lectura_abierto.readlines()
            #print(f"Contenido: {contenido}")
            for linea in contenido:
                print(linea)                       
            archivo_lectura_abierto.close()
        except Exception as error:
            print("Error")
        
    def actualizar(self):
        try:
            contador = 0
            buscador = (input("Ingrese la ciudad del cine deseado: "))
            buscador = (buscador + "\n")
            path = "./cines.txt"
            archivo_lectura_abierto = open(path)
            contenido = archivo_lectura_abierto.readlines()
#            print(f"buscador: {buscador}")
            for linea in contenido:
#                print(contador)
#                print(linea)
#                print("----")
                if (linea == buscador):
                    aux = contador
#                    print(aux)
                    print("Cine a modificar:\n")
                    for aux in range (contador, contador+4):
                        print(contenido[aux])
                    respuesta = (input("¿Desea actualizar este cine? (s/n)"))
                    if(respuesta == "s"):
                        print("Ingrese la Ciudad, Nombre, Sala y Precio del cine a actualizar.")
                        aux = contador
                        for aux in range(contador, contador+4):
#                            print(f"aux: {aux}")
                            nuevo = (input("Nuevo dato: "))
                            contenido[aux] = (nuevo + "\n")
                    
                contador += 1
                        
            archivo_escritura_abierto = open(path, mode = "w")
            archivo_escritura_abierto.writelines(contenido)
            archivo_escritura_abierto.close()
            archivo_lectura_abierto.close()
        except Exception as error:
            print("Error")
            
    def eliminar(self):
        try:
            contador = 0
            buscador = (input("Ingrese la ciudad del cine a eliminar: "))
            buscador = (buscador + "\n")
            path = "./cines.txt"
            archivo_lectura_abierto = open(path)
            contenido = archivo_lectura_abierto.readlines()
#            print(f"buscador: {buscador}")
            for linea in contenido:
                if (linea == buscador):
                    aux = contador
#                    print(aux)
                    print("Cine a eliminar:\n")
                    for aux in range (contador, contador+4):
                        print(contenido[aux])
                    respuesta = (input("¿Desea eliminar este cine? (s/n)"))
                    if(respuesta == "s"):
                        aux = contador
                        for aux in range(contador, contador+4):
                            aux2 = (aux+5)
                            #print(contenido)
                            contenido[aux] = contenido[aux2]
                            contenido[aux2] = ""
                    
                contador += 1
            
            print(contenido)
            archivo_escritura_abierto = open(path, mode = "w")
            archivo_escritura_abierto.writelines(contenido)
            archivo_escritura_abierto.close()
            archivo_lectura_abierto.close()
        except Exception as error:
            print("Error")
```

```python
class Pelicula:
    titulo = None
    anio = None
    genero = None
    duracion = 0
          
    def asignar_cine(self):
        try:
            contador = 0
            buscador = (input("Ingrese la ciudad del cine donde se proyectará esta película: "))
            buscador = (buscador + "\n")
            #print(f"buscador: {buscador}")
            path = "./cines.txt"
            print("a")
            path_cartelera = "./cartelera.txt"       
            print("b")
            archivo_lectura_abierto = open(path)
            print("c")
            archivo_cartelera = open(path_cartelera, mode = "a")
            print("d")
            contenido = archivo_lectura_abierto.readlines()
            print("e")
            contenido_cartelera = archivo_cartelera.readlines()
            print("aaaaaa")
            for linea in contenido:
#                print(contador)
#                print(linea)
#                print("----")
                if (linea == buscador):
                    aux = contador
    #                    print(aux)
                    print("Cine elegido:\n")
                    for aux in range (contador, contador+4):
                        print(contenido[aux])
                    respuesta = (input("¿Desea añadir la película a este cine? (s/n)"))
                    aux = contador
                    if(respuesta == "s"):
                        for aux in range (contador, contador+4):
                            contenido_cartelera.writelines(contenido[aux])
                    
            contador += 1
            archivo_lectura_abierto.close()
            archivo_cartelera.close()
        except Exception as error:
            print("Error")

        
    def guardar(self):
        try:
            self.titulo = (input("Título: "))
            self.anio = (input("Año: "))
            self.genero = (input("Genero: "))
            self.duracion = (input("Duración: "))
            
            path = "./peliculas.txt"       
            archivo_escritura_abierto = open(path, mode = "a")
            archivo_escritura_abierto.writelines([f"{self.titulo}\n", 
                                                          f"{self.anio}\n",
                                                          f"{self.genero}\n",
                                                          f"{self.duracion}\n\n"])
            #self.asignar_cine()
            path_cartelera = "./cartelera.txt"       
            archivo_cartelera = open(path_cartelera, mode = "a")
            archivo_cartelera.writelines([f"{self.titulo}\n", 
                                          f"{self.anio}\n",
                                          f"{self.genero}\n",
                                          f"{self.duracion}\n\n"])
            archivo_escritura_abierto.close()
            archivo_cartelera.close()
        except Exception as error:
            print("Error")
        return "Película Guardada Exitosamente"
        
    def leer(self):
        try:
            path = "./peliculas.txt"
            archivo_lectura_abierto = open(path)
            contenido = archivo_lectura_abierto.readlines()
            #print(f"Contenido: {contenido}")
            for linea in contenido:
                print(linea)                       
            archivo_lectura_abierto.close()
        except Exception as error:
            print("Error")
        
    def actualizar(self):
        try:
            contador = 0
            buscador = (input("Ingrese el título de la película deseada: "))
            buscador = (buscador + "\n")
            path = "./peliculas.txt"
            archivo_lectura_abierto = open(path)
            contenido = archivo_lectura_abierto.readlines()
#            print(f"buscador: {buscador}")
            for linea in contenido:
#                print(contador)
#                print(linea)
#                print("----")
                if (linea == buscador):
                    aux = contador
#                    print(aux)
                    print("Película a modificar:\n")
                    for aux in range (contador, contador+4):
                        print(contenido[aux])
                    respuesta = (input("¿Desea actualizar esta película? (s/n)"))
                    if(respuesta == "s"):
                        print("Ingrese el Título, Año, Género y Duración de la película a actualizar.")
                        aux = contador
                        for aux in range(contador, contador+4):
#                            print(f"aux: {aux}")
                            nuevo = (input("Nuevo dato: "))
                            contenido[aux] = (nuevo + "\n")
                    
                contador += 1
                        
            archivo_escritura_abierto = open(path, mode = "w")
            archivo_escritura_abierto.writelines(contenido)
            archivo_escritura_abierto.close()
            archivo_lectura_abierto.close()
        except Exception as error:
            print("Error")
            
    def eliminar(self):
        try:
            contador = 0
            buscador = (input("Ingrese el titulo de la película a eliminar: "))
            buscador = (buscador + "\n")
            path = "./peliculas.txt"
            archivo_lectura_abierto = open(path)
            contenido = archivo_lectura_abierto.readlines()
#            print(f"buscador: {buscador}")
            for linea in contenido:
                if (linea == buscador):
                    aux = contador
#                    print(aux)
                    print("Película a eliminar:\n")
                    for aux in range (contador, contador+4):
                        print(contenido[aux])
                    respuesta = (input("¿Desea eliminar esta película? (s/n)"))
                    if(respuesta == "s"):
                        aux = contador
                        for aux in range(contador, contador+4):
                            aux2 = (aux+4)
                            contenido[aux] = contenido[aux2]
                            contenido[aux2] = ""
                    
                contador += 1
                        
            archivo_escritura_abierto = open(path, mode = "w")
            archivo_escritura_abierto.writelines(contenido)
            archivo_escritura_abierto.close()
            archivo_lectura_abierto.close()
        except Exception as error:
            print("Error")
```

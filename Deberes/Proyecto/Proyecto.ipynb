```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
path = 'C://Users//MORALES//Documents//GitHub//py-morales-banda-juan-jose//Deberes//Proyecto//Anime.csv'
columnas = ['animeID',' name',' type',' source',' producers',' genre',' studio',' episodes',' status',' rating',' score',' premiered']
columnas2 = ['name','type','source','producers','genre','studio','episodes','status','rating','score','premiered']
df1 = pd.read_csv(path, usecols=columnas, index_col = 'animeID')
df1.columns = columnas2
```

```python
df1['rating'].value_counts().plot(kind='bar',legend='Reverse')
```

```python
df1['type'].value_counts().plot(kind='pie',cmap ='Paired')
plt.ylabel('')
```

```python
#df1['episodes'].sort_values().tail(20).plot(kind='bar',legend='Reverse')
#df1['episodes'].sort_values(ascending = False).head(20).plot(kind='bar',legend='Reverse')
df_columnas = ['name','episodes']
valores = df1[df_columnas].sort_values(by='episodes', ascending = False).head(20)
valores.index = valores['name']
valores.plot( kind='bar')
```

```python
df1['status'].value_counts().plot(kind='barh',legend='Reverse')
```

```python
df1.source.value_counts().plot(kind='bar')
```

```python
#df1['score'].sort_values().tail(20).plot(kind='bar',legend='Reverse')
#df1['score'].sort_values(ascending = False).head(20).plot(kind='bar',legend='Reverse')
df_columnas = ['name','score']
valores = df1[df_columnas].sort_values(by='score', ascending = False).head(20)
valores.index = valores['name']
valores.plot( kind='bar')
```

```python
pivote = pd.pivot_table(df1, index=['status', 'rating'], values=['source'], columns=['type'], aggfunc='count')
pivote.groupby('status').plot(kind = 'bar', legend = 'Reverse')
```

```python
estudios = df1['studio'].value_counts().head(20)
estudios[1:19].plot(kind='bar')
```

```python
df1['premiered'].value_counts().head(20).plot(kind='bar')
```
